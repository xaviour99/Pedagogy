```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pickle
from sklearn.linear_model import LinearRegression
```

```python
#reading the  datasets for diabetes 
data=pd.read_csv("SAMPLE.csv")
```

```python
data['TOTAL IA'].plot.hist(rwidth=0.8)
plt.title("Student Internal Mark")
plt.xlabel("Internal")
plt.ylabel("no of students")
plt.rcParams['figure.figsize'] = [10,4]
plt.show()
```

```python
data['TOTAL(A)'].plot.hist(rwidth=0.8)
plt.title("Student assignment Mark")
plt.xlabel("Assignment")
plt.ylabel("no of students")
plt.rcParams['figure.figsize'] = [10,4]
plt.show()
```

```python
data['TOTAL(Q)'].plot.hist(rwidth=0.8)
plt.title("Student assignment Mark")
plt.xlabel("Assignment")
plt.ylabel("no of students")
plt.rcParams['figure.figsize'] = [10,4]
plt.show()
```

```python
data['LAB (10)'].plot.hist(rwidth=0.8)
plt.title("Student assignment Mark")
plt.xlabel("Assignment")
plt.ylabel("no of students")
plt.rcParams['figure.figsize'] = [10,4]
plt.show()
```

```python
num = pd.value_counts(data['SEM_RESULT'],sort=True).sort_index()
num.plot(kind='bar')
plt.title('histogram')
plt.xlabel('RESULT')
plt.ylabel('number')
```

```python
x = data[['TOTAL IA','TOTAL(A)','TOTAL(Q)','LAB (10)']]
y = data['SEM_MARKS']
```

```python
print(x.shape)
print(y.shape)
```

```python
x.head()
```

```python
lg_model=LinearRegression()
```

```python
lg_model.fit(x,y)
```

```python
lg_model.predict([[11.00,5.0,2.00,0]])
```

```python
with open("model.pkl","wb") as f:
    pickle.dump(lg_model,f)
```
